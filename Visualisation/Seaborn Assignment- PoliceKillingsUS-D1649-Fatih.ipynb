## 2: Plot 12 most dangerous cities

```python
most_dangerous_cities = kill['city'].value_counts().iloc[0:12]
type(most_dangerous_cities)
```

```python
fgr = plt.subplots(figsize = (8,6))
plt.xticks(rotation = 45)
plt.title("Most dangerous cities", fontdict = {'fontsize': 14,        
        'color': 'Blue'})
sns.barplot(y= most_dangerous_cities.values, x = most_dangerous_cities.index);
```

```python
fgr = plt.subplots(figsize = (8,6))
plt.xticks(rotation = 45)
plt.title("Most dangerous cities", fontdict = {'fontsize': 14,        
        'color': 'Blue'})
sns.countplot(kill.city, order=pd.value_counts(kill.city).iloc[:12].index);
```

## 3: Plot 20 most dangerous states

```python
most_dangerous_20_states = kill['state'].value_counts().iloc[0:20]
most_dangerous_20_states
```

```python
index = most_dangerous_20_states.index
values = most_dangerous_20_states.values
fgr = plt.subplots(figsize = (8,6))
plt.xticks(rotation = 45)
plt.title("Most dangerous states", fontdict = {'fontsize': 14,        
        'color': 'Blue'})
sns.barplot(index, values);
```

```python
fgr = plt.subplots(figsize = (8,6))
plt.xticks(rotation = 45)
plt.title("Most dangerous states", fontdict = {'fontsize': 14,        
        'color': 'Blue'})
sns.countplot(kill.state, order=pd.value_counts(kill.state).iloc[:20].index);
```
